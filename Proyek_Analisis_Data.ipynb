#### melihat nilai outlier

```python
# deteksi outlier dengan metode IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# outlier pada kolom PM2.5 di setiap distrik
for district, df in zip(['Aotizhongxin', 'Changping', 'Dongsi', 'Shunyi', 'Tiantan'],
                        [df_aotizhongxin, df_changping, df_dongsi, df_shunyi, df_tiantan]):
    for column in ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO','O3']:
        outliers = detect_outliers_iqr(df, column)
        print(f"Outlier di Distrik {district} pada kolom {column}: {len(outliers)} data")
```

#### **Menghapus Outlier**

- 1. metode pertama menghapus outlier tanpa memberi nilai baru

```python
# fungsi untuk mengganti outlier dengan nilai lower_bound atau upper_bound
def replace_outliers(df, columns):
    df_new = df.copy()
    for column in columns:
        if column in df_new.columns:
            Q1 = df_new[column].quantile(0.25)
            Q3 = df_new[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            df_new.loc[df_new[column] < lower_bound, column] = lower_bound
            df_new.loc[df_new[column] > upper_bound, column] = upper_bound
        else:
            print(f"Kolom '{column}' tidak ditemukan di dataframe.")
    return df_new

# Ganti nilai outlier pada setiap dataframe
df_aotizhongxin_new = replace_outliers(df_aotizhongxin, ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3'])
df_changping_new = replace_outliers(df_changping, ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3'])
df_dongsi_new = replace_outliers(df_dongsi, ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3'])
df_shunyi_new = replace_outliers(df_shunyi, ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3'])
df_tiantan_new = replace_outliers(df_tiantan, ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3'])
```

- 2. mengecek outlier yang telah dihapus

```python
# deteksi outlier dengan metode IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# outlier pada kolom PM2.5 di setiap distrik
for district, df in zip(['Aotizhongxin', 'Changping', 'Dongsi', 'Shunyi', 'Tiantan'],
                        [df_aotizhongxin_new, df_changping_new, df_dongsi_new, df_shunyi_new, df_tiantan_new]):
    for column in ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO','O3']:
        outliers = detect_outliers_iqr(df, column)
        print(f"Outlier di Distrik {district} pada kolom {column}: {len(outliers)} data")
```

```python
def get_season(month):
    if month in [12, 1, 2]: #Musim Dingin
        return 'Winter'
    elif month in [3, 4, 5]: # Musimg Semi
        return 'Spring'
    elif month in [6, 7, 8]: # Musim Panas
        return 'Summer'
    else:
        return 'Fall' # Musim Gugur bulan (9, 10 ,11)

#tambah kolom musim berdasarkan bulan
for df in [df_aotizhongxin_new, df_changping_new, df_dongsi_new, df_shunyi_new, df_tiantan_new]:
    df['season'] = df['month'].apply(get_season)

#menggabungkan dataframe menjadi satu
seasonal_pm25 = pd.DataFrame({
    'Aotizhongxin': df_aotizhongxin_new.groupby('season')['PM2.5'].mean(),
    'Changping': df_changping_new.groupby('season')['PM2.5'].mean(),
    'Dongsi': df_dongsi_new.groupby('season')['PM2.5'].mean(),
    'Shunyi': df_shunyi_new.groupby('season')['PM2.5'].mean(),
    'Tiantan': df_tiantan_new.groupby('season')['PM2.5'].mean()
})

#Visualisasi rata-rata PM2.5 berdasarkan musim
seasonal_pm25.plot(kind='bar', figsize=(10, 6), rot=0)
plt.title('Rata-rata PM2.5 Berdasarkan Musim di Lima Distrik Utama', fontsize=14)
plt.xlabel('Musim', fontsize=12)
plt.ylabel('Rata-rata PM2.5', fontsize=12)
plt.legend(title='Distrik', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
def get_season(month):
    if month in [12, 1, 2]: #Musim Dingin
        return 'Winter'
    elif month in [3, 4, 5]: # Musimg Semi
        return 'Spring'
    elif month in [6, 7, 8]: # Musim Panas
        return 'Summer'
    else:
        return 'Fall' # Musim Gugur (9, 10 ,11)

for df in [df_aotizhongxin_new, df_changping_new, df_dongsi_new, df_shunyi_new, df_tiantan_new]:
    df['season'] = df['month'].apply(get_season)

polutans = ['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3']
seasonal_data = {}

# hitung rata-rata untuk setiap polutan dan musim
for polutan in polutans:
    seasonal_data[polutan] = pd.DataFrame({
        'Aotizhongxin': df_aotizhongxin_new.groupby('season')[polutan].mean(),
        'Changping': df_changping_new.groupby('season')[polutan].mean(),
        'Dongsi': df_dongsi_new.groupby('season')[polutan].mean(),
        'Shunyi': df_shunyi_new.groupby('season')[polutan].mean(),
        'Tiantan': df_tiantan_new.groupby('season')[polutan].mean()
    })

# visualisasi untuk setiap polutan
fig, axes = plt.subplots(len(polutans), 1, figsize=(12, 5 * len(polutans)), sharex=True)

for ax, polutan in zip(axes, polutans):
    seasonal_data[polutan].plot(kind='bar', ax=ax)
    ax.set_title(f'Rata-rata {polutan} Berdasarkan musim di lima distrik utama', fontsize=14)
    ax.set_ylabel(f'Rata-rata {polutan}', fontsize=12)
    ax.legend(title='Distrik', loc='upper right')
    ax.set_xticklabels([])
    ax.grid(True)

    # tambah nama musim di tengah setiap bar
    for i in range(len(seasonal_data[polutan])):
        ax.text(i, seasonal_data[polutan].max().max() * -0.1, seasonal_data[polutan].index[i],
                ha='center', va='bottom', fontsize=10, color='black')

plt.xlabel('Musim', fontsize=10)
plt.tight_layout()
plt.show()
```

## Analisis Lanjutan Dengan Clustering Analysis

####  **Membuat rata-rata untuk menentukan cluster**

1. Metode Manual Groupping

```python
def seasonal_avg(df, district_name):
    return df.groupby(['year', 'season']).agg({
        'PM2.5': 'mean',
        'PM10': 'mean',
        'SO2': 'mean',
        'NO2': 'mean',
        'CO': 'mean',
        'O3': 'mean'
    }).reset_index().assign(District=district_name)
```

```python
df_aotizhongxin_avg = seasonal_avg(df_aotizhongxin_new, 'Aotizhongxin')
df_changping_avg = seasonal_avg(df_changping_new, 'Changping')
df_dongsi_avg = seasonal_avg(df_dongsi_new, 'Dongsi')
df_shunyi_avg = seasonal_avg(df_shunyi_new, 'Shunyi')
df_tiantan_avg = seasonal_avg(df_tiantan_new, 'Tiantan')
```

```python
df_avg_musim = pd.concat([df_aotizhongxin_avg, df_changping_avg, df_dongsi_avg, df_shunyi_avg, df_tiantan_avg])
```

```python
df_avg_musim.to_csv('df_avg_musim.csv', index=False)
```

Pernyataan apabila nilai sebelum dilakukan mean PM2.5 > dari nilai PM2.5 setelah dilakukan rata-rata dan nilai awal PM10 > median PM10

```python
df_avg_musim['Cluster'] = np.where((df_avg_musim['PM2.5'] > df_avg_musim['PM2.5'].mean()) &
                                              (df_avg_musim['PM10'] > df_avg_musim['PM10'].median()),
                                              'Tinggi Polusi', 'Rendah Polusi')
print(df_avg_musim[['year', 'season', 'District', 'Cluster']])
```
